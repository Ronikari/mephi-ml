### Задание 2
Найти безусловные экстремумы функции $ u = 3x^3 + y^2 + z^2 + 6xy - 2z + 1 $.

**Решение:**  

Найти частные производные функции по $ x $, $ y $, и $ z $:  

- $ \frac{\partial u}{\partial x} = 9x^2 + 6y $

- $ \frac{\partial u}{\partial y} = 2y + 6x $

- $ \frac{\partial u}{\partial z} = 2z - 2 $

Приравнять каждую из частных производных к нулю и решить полученную систему уравнений:

$\left\{\begin{matrix} 9x^2 + 6y = 0 \\ 2y + 6x = 0 \\ 2z - 2 = 0 \end{matrix}\right.$

$\left\{\begin{matrix} 9x^2 + 6y = 0 \\ y + 3x = 0 \\ z = 1 \end{matrix}\right.$

$\left\{\begin{matrix} y = - 3x \\ z = 1 \\ 9x^2 - 18x = 0 \end{matrix}\right.$

$\left\{\begin{matrix} y = - 3x \\ z = 1 \\ x(x - 2) = 0 \end{matrix}\right.$


Найти $ y $ для каждого значения $ x $:
- При $ x = 0 $:

  $ y = -3(0) = 0 $
- При $ x = 2 $:

  $ y = -3(2) = -6 $

Получить в конечном итоге критические точки:
- $ (0, 0, 1) $

- $ (2, -6, 1) $

---

Найти частные производные второго порядка:

$\left\{\begin{matrix} \frac{\partial^2 u}{\partial x^2} = 18x \\ \frac{\partial^2 u}{\partial y^2} = 2 \\ \frac{\partial^2 u}{\partial z^2} = 2 \end{matrix}\right.$

$\left\{\begin{matrix} \frac{\partial^2 u}{\partial x \partial y} = 6 \\ \frac{\partial^2 u}{\partial y \partial z} = 0 \\ \frac{\partial^2 u}{\partial z \partial x} = 0 \end{matrix}\right.$

Составить матрицу Гессе $ H $:

$$
H = \begin{pmatrix}
\frac{\partial^2 u}{\partial x^2} & \frac{\partial^2 u}{\partial x \partial y} & \frac{\partial^2 u}{\partial x \partial z} \\
\frac{\partial^2 u}{\partial y \partial x} & \frac{\partial^2 u}{\partial y^2} & \frac{\partial^2 u}{\partial y \partial z} \\
\frac{\partial^2 u}{\partial z \partial x} & \frac{\partial^2 u}{\partial z \partial y} & \frac{\partial^2 u}{\partial z^2}
\end{pmatrix}
=
\begin{pmatrix}
18x & 6 & 0 \\
6 & 2 & 0 \\
0 & 0 & 2
\end{pmatrix} 
$$

---

Матрица Гессе для точки экстремума $ (0, 0, 1) $ (при $ x = 0 $):

$
H(0, 0, 1) = \begin{pmatrix}
0 & 6 & 0 \\
6 & 2 & 0 \\
0 & 0 & 2
\end{pmatrix}
$

Найти детерминант полученной матрицы $ H $:

$ \det(H) = (0 \cdot 2) - (6 \cdot 6) = -36 $
   
**$ (0, 0, 1) $ является седловой точкой и не является экстремумом.**

---

Матрица Гессе для точки экстремума $ (2, -6, 1) $ (при $ x = 2 $):

$
H(2, -6, 1) = \begin{pmatrix}
36 & 6 & 0 \\
6 & 2 & 0 \\
0 & 0 & 2
\end{pmatrix}
$

Найти детерминант полученной матрицы $ H $:

$ \det(H) = 36(2 \cdot 2) - 6(6 \cdot 0) + 0 = 144 $
   
**$ (2, -6, 1) $ является локальным минимумом.**

---

**Ответ:** функция $ u = 3x^3 + y^2 + z^2 + 6xy - 2z + 1 $ имеет локальный минимум в точке $ (2, -6, 1) $.

### Задание 4
Вам предоставлены данные о людях с наличием или отсутствием заболевания.  
  
Признаки:  

- `Age` — возрастная группа.
- `Test` — позитивный или негативный тест на заболевание.
- `Status` — целевая переменная, есть инфекция или нет.
Необходимо реализовать алгоритм Наивного Байеса для решения задачи классификации. Обязательно оцените качество полученного результата по итогу.

```python
# Загрузить исходные данные и сформировать датасет
df = pd.read_csv("C:/Users/ludkov-as/repo/math_analysis/exam.csv")
```

```python
# Изменить тип данных столбцов
df['Age_Group'] = df['Age_Group'].astype('category').cat.codes
df['Test'] = df['Test'].astype('category').cat.codes
df['Status'] = df['Status'].astype('category').cat.codes

# Создать матрицу наблюдений Х
X = df[['Age_Group', 'Test']]

# Создать вектор правильных ответов у
y = df['Status']

# Подготовить тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Обучить модели с использованием НБК
model = GaussianNB()
model.fit(X_train, y_train)

# Осуществить прогноз на тестовой выборке
y_pred = model.predict(X_test)

# Осуществить оценку качества модели
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f'Точность модели: {accuracy:.2f}')
print('Матрица ошибок:\n', conf_matrix)
print('Интерпретация полученных результатов:\n', class_report)
```

### Задание 5
Дана функция:  
  
$f(x) = x^4 + 3x^3 - 12x^2 +7x - 2$  

Оптимизируйте её с помощью метода Ньютона, используя язык программирования Python.  

За начальную точку возьмите $x = 10$, необходимую точность $0.0001$.  

```python
# Начальная точка и необходимая точность
initial_x = 10
tol = 0.0001

# Инициализация функции
def f(x):
    return x**4 + 3*x**3 - 12*x**2 + 7*x - 2

# Функция для нахождения производной первого порядка
def f_prime(x):
    return 4*x**3 + 9*x**2 - 24*x + 7

# Функция для нахождения производной второго порядка
def f_prime_2(x):
    return 12*x**2 + 18*x - 24

# Функция для реализации метода Ньютона
def newtons_method(x0, tol):
    x_curr = x0
    while True:
        f_prime_val = f_prime(x_curr)
        f_double_prime_val = f_prime_2(x_curr)
        
        # Если производная второго порядка равна нулю, остановить выполнение функции
        if f_double_prime_val == 0:
            print("Метод Ньютона неприменим (f'' = 0)")
            return None
        
        # Актуализировать значение x
        x_curr1 = x_curr - f_prime_val / f_double_prime_val
        
        # Проверить на сходимость
        if abs(x_curr1 - x_curr) < tol:
            break
        x_curr = x_curr1 
    return x_curr

# Инициализировать выполнение метода Ньютона
x_opt = newtons_method(initial_x, tol)
print(f"Оптимальное значение x: {x_opt}")
print(f"Значение функции в оптимальной точке: {f(x_opt)}")
```

### Задание 6
Дано распределение случайной величины, которая отражает вероятность получения разного количества спам-писем в течение дня:   
|  0  |  1  |  2  |   3  |   4  |   5  |  6  |  7  |
|:---:|:---:|:---:|:----:|:----:|:----:|:---:|:---:|
| 0.1 | 0.3 | 0.1 | 0.25 | 0.05 | 0.05 | 0.1 |  ?  |

1. Найдите вероятность получения 7 спам-писем.
2. Найдите математическое ожидание для количества полученных писем.
3. Найдите дисперсию для количества полученных писем. Ответ округлите до сотых.

```python
X = [0.1, 0.3, 0.1, 0.25, 0.05, 0.05, 0.1]

# Найти вероятность получения 7 спам-писем
n_sum = sum(X)
p = 1 - n_sum

# Добавить полученную вероятность в список (вместо недостающего значения в таблице)
X.append(p)
print(f"Вероятность получения 7 спам-писем: {p * 100:.0f}%")

# Вычислить математическое ожидание
E_x = 0
for i in range(len(X)):
    E_x += i * X[i]
print(f"Математическое ожидание (E(x)): {E_x:.2f}")

# Вычислить дисперсию, но сначала найти E(X^2)
E_x2 = 0
for i in range(len(X)):
    E_x2 += (i ** 2) * X[i]
D_x = E_x2 - (E_x ** 2)
print(f"Дисперсия (D(x)): {D_x:.2f}")
```

### Задание 7
За 8 часов рабочего дня в колл-центр поступает в среднем 16 звонков. Найдите вероятность, что за час в колл-центр поступит не более 4 и не менее 2 звонков. Ответ округлите до тысячных.
  
**Решение:**  
  
Так как в условии задачи подразумевается поиск вероятности возникновения $k$ событий за период времени, следует использовать распределение Пуассона.  
  
Распределение Пуассона считается по следующей формуле:  
  
$ P(X = k) = \frac{\lambda^k e^{- \lambda}}{k!} $  
  
где:
- $k$ — число событий;
- $\lambda$ — среднее число событий.

```python
# Ожидаемое число событий за период времени
λ = 2

# Так как по условию задачи не используются готовые функции,
# реализовать функцию нахождения факториала самостоятельно
def factorial(k):
    result = 1
    for i in range(1, k + 1):
        result *= i
    return result

# Вычислить вероятности для k = 2, 3, 4 по распределению Пуассона
probs = [(λ**k * math.exp(-λ)) / factorial(k) for k in range(2, 5)]

# Суммарная вероятность
prob_sum = round(sum(probs), 3)

print(prob_sum)
```

–ó–∞–¥–∞–Ω–∏–µ 1

–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã (—á–µ—Ä–µ–∑ —Ñ—É–Ω–∫—Ü–∏—é input()) –≤–≤–æ–¥–∏—Ç —á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é –∏ –ø—Ä–æ–±–µ–ª –¥–≤–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —á–∏—Å–µ–ª (–Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä—ã —Ç–æ–≤–∞—Ä–æ–≤/—É—Å–ª—É–≥, –ø—Ä–æ—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–º 'A' –∏ –∫–ª–∏–µ–Ω—Ç–æ–º 'B' –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ ¬´–ê–≤–∏—Ç–æ¬ª).
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–∞–∂–¥—ã–π –≤–≤–æ–¥ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π —Å–ø–∏—Å–æ–∫, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, –∞ —Ç–∞–∫–∂–µ —Å–æ—Å—Ç–∞–≤–∏—Ç—å –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –¥–≤—É—Ö –∏—Å—Ö–æ–¥–Ω—ã—Ö. –î—É–±–ª–∏–∫–∞—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —É–¥–∞–ª–∏—Ç—å, –∞ —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∏–π —Å–ø–∏—Å–æ–∫ ‚Äî –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω.
–í–∞–∂–Ω–æ! –ï—Å–ª–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ –ø—É—Å—Ç–æ–µ, —Ç–æ –Ω–∞ —ç–∫—Ä–∞–Ω –¥–æ–ª–∂–µ–Ω –≤—ã–≤–æ–¥–∏—Ç—å—Å—è –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫.
–ï—Å–ª–∏ –≤–≤–æ–¥ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –≤ —Å–ø–∏—Å–æ–∫ –∏–∑ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, –≤–æ–∑–Ω–∏–∫–∞–µ—Ç –æ—à–∏–±–∫–∞. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –æ—Ç–ª–∞–≤–ª–∏–≤–∞—Ç—å —ç—Ç—É –æ—à–∏–±–∫—É –∏ –≤—ã–≤–æ–¥–∏—Ç—å –Ω–∞ —ç–∫—Ä–∞–Ω —Ñ—Ä–∞–∑—É –ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥.
–î–ª—è –ø–æ–∏—Å–∫–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–π –≤ —Å–ø–∏—Å–∫–∞—Ö –Ω–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é pure_intersection(), –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å —Å–ø–∏—Å–æ–∫ —Å —É–Ω–∏–∫–∞–ª—å–Ω—ã–º–∏ —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è.

```python
input_string1 = input('–í–≤–µ–¥–∏—Ç–µ 1-—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä–æ–≤: ')
input_string2 = input('–í–≤–µ–¥–∏—Ç–µ 2-—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä–æ–≤: ')
# –≤–∞—à –∫–æ–¥ –∑–¥–µ—Å—å
def pure_intersection(s1, s2):
    # –ü–æ–ø—ã—Ç–∞–µ–º—Å—è —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å —Å–ø–∏—Å–∫–∏ –∏–∑ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª
    try:
        # –£–¥–∞–ª–∏–º –∏–∑ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–∏ –∑–∞–ø—è—Ç—ã–µ, —Ä–∞–∑–¥–µ–ª–∏–º —Å—Ç—Ä–æ–∫—É –∏ –∑–∞–ø–∏—à–µ–º –≤ —Å–ø–∏—Å–æ–∫
        list1 = list(s1.replace(',', '').split(' '))
        # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å—Ç—Ä–æ–∫–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞
        list1 = [int(x) for x in list1]
        # –ü—Ä–æ–≤–µ–¥—ë–º –∞–Ω–∞–ª–æ–≥–∏—á–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è –¥–ª—è –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–∏
        list2 = list(s2.replace(',', '').split(' '))
        list2 = [int(x) for x in list2]
    except ValueError:
        # –ò—Å–∫–ª—é—á–µ–Ω–∏–µ –≤–æ–∑–Ω–∏–∫–∞–µ—Ç, –µ—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤—ë–ª –Ω–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞
        print('–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥')
        return
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å–ø–∏—Å–∫–∏ –≤ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –∏ –Ω–∞–π–¥—ë–º –∏—Ö –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∫–æ–Ω–µ—á–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—Ä–∞—Ç–Ω–æ –≤ —Å–ø–∏—Å–æ–∫
    result = list(set(list1).intersection(set(list2)))
    print(result)

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é
pure_intersection(input_string1, input_string2)
```

–ó–∞–¥–∞–Ω–∏–µ 2

–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø–æ–¥–∞—ë—Ç –Ω–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞ —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã (—á–µ—Ä–µ–∑ —Ñ—É–Ω–∫—Ü–∏—é input()).
–û–Ω –ø—Ä–∏–≤—ã–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è Excel, –ø–æ—ç—Ç–æ–º—É –æ—Ç–¥–µ–ª—è–µ—Ç –¥—Ä–æ–±–Ω—É—é —á–∞—Å—Ç—å —á–∏—Å–ª–∞ –∑–∞–ø—è—Ç–æ–π, –∞ –Ω–µ —Ç–æ—á–∫–æ–π. –ö—Ä–æ–º–µ —Ç–æ–≥–æ, –æ–Ω –º–æ–∂–µ—Ç —Å–ª—É—á–∞–π–Ω–æ –≤–≤–µ—Å—Ç–∏ —á–∏—Å–ª–æ ¬´—Å–ª–æ–≤–∞–º–∏¬ª (–Ω–∞–ø—Ä–∏–º–µ—Ä, –Ω–µ "5", –∞ ¬´–ø—è—Ç—å¬ª) ‚Äî –≤ —ç—Ç–æ–º —Å–ª—É—á–∞–µ —Å–ª–æ–≤–æ –Ω–∞–¥–æ –∏–≥–Ω–æ—Ä–∏—Ä–æ–≤–∞—Ç—å.
–í–∞—à–∞ –∑–∞–¥–∞—á–∞ ‚Äî –≤—ã–≤–µ—Å—Ç–∏ –º–∏–Ω–∏–º—É–º –∏ –º–∞–∫—Å–∏–º—É–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º –±—É–¥–µ—Ç –≤–≤–µ–¥–µ–Ω–æ —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ —á–∏—Å–ª–æ.
–î–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∏–Ω–∏–º—É–º–∞ –∏ –º–∞–∫—Å–∏–º—É–º–∞ –Ω–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é find_min_max, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å –¥–≤–∞ —á–∏—Å–ª–∞.

```python
input_string = input('–í–≤–µ–¥–∏—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª: ')
# –≤–∞—à –∫–æ–¥ –∑–¥–µ—Å—å
def find_min_max(s):
    # –†–∞–∑–¥–µ–ª–∏–º —Å—Ç—Ä–æ–∫—É –∏ —Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫
    lst = list(s.split(' '))
    try:
        # –ü–æ–ø—ã—Ç–∞–µ–º—Å—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å–ø–∏—Å–∫–∞ –≤ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞
        lst = [float(x) for x in lst]
    except ValueError:
        # –ï—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤—ë–ª —á–∏—Å–ª–∞ —Å –∑–∞–ø—è—Ç–æ–π –≤ –∫–∞—á–µ—Å—Ç–≤–µ —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—è,
        # –∑–∞–º–µ–Ω–∏–º –Ω–∞ —Ç–æ—á–∫—É –∏ –¥–æ–±–∞–≤–∏–º –≤ —Å–ø–∏—Å–æ–∫
        # –ï—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤—ë–ª –∑–Ω–∞—á–µ–Ω–∏–µ, —Å–æ—Å—Ç–æ—è—â–µ–µ —Ç–æ–ª—å–∫–æ –∏–∑ –±—É–∫–≤ - –ø—Ä–æ–∏–≥–Ω–æ—Ä–∏—Ä—É–µ–º
        lst = [float(x.replace(',', '.')) for x in lst if not x.isalpha()]
    # –í—ã—á–∏—Å–ª–∏–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç—ã —Å–ø–∏—Å–∫–∞
    print(f"Minimum: {min(lst)}")
    print(f"Maximum: {max(lst)}")

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é
find_min_max(input_string)
```

–ó–∞–¥–∞–Ω–∏–µ 3

–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã (—á–µ—Ä–µ–∑ —Ñ—É–Ω–∫—Ü–∏—é input()) –≤–≤–æ–¥–∏—Ç —á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é –∏ –ø—Ä–æ–±–µ–ª –Ω–µ—É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª.
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –≤–≤–æ–¥ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ —Å–ø–∏—Å–æ–∫ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª –∏ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω –º–µ–¥–∏–∞–Ω—É —ç—Ç–æ–π —á–∏—Å–ª–æ–≤–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏.
–î–ª—è –ø–æ–∏—Å–∫–∞ –º–µ–¥–∏–∞–Ω—ã –≤ —Å–ø–∏—Å–∫–µ –Ω–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é find_median(), –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ. –†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–∞–±–æ—Ç—ã —Ñ—É–Ω–∫—Ü–∏–∏ –Ω—É–∂–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞ —ç–∫—Ä–∞–Ω.
–ï—Å–ª–∏ –ø–µ—Ä–µ–¥–∞–Ω–Ω–∞—è –≤ —Ñ—É–Ω–∫—Ü–∏—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –æ–∫–∞–∂–µ—Ç—Å—è –ø—É—Å—Ç–æ–π (–ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –Ω–∏—á–µ–≥–æ –Ω–µ –≤–≤—ë–ª) –∏–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤—ë–ª —á–∏—Å–ª–∞ –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ (–≤–≤–æ–¥ –Ω–µ–ª—å–∑—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –≤ —á–∏—Å–ª–∞), –Ω–∞ —ç–∫—Ä–∞–Ω –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –≤—ã–≤–µ–¥–µ–Ω–∞ —Ñ—Ä–∞–∑–∞ –ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥.

```python
input_string = input('–í–≤–µ–¥–∏—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª: ')
# –≤–∞—à –∫–æ–¥ –∑–¥–µ—Å—å
def find_median(s):
    # –†–∞–∑–¥–µ–ª–∏–º —Å—Ç—Ä–æ–∫—É –∏ —Å–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–æ–∫
    lst = list(s.replace(',', '').split(' '))
    try:
        # –ü–æ–ø—ã—Ç–∞–µ–º—Å—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã —Å–ø–∏—Å–∫–∞ –≤ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞
        # –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é
        # –í –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–±—Ä–∞—Å—ã–≤–∞–µ–º –∏—Å–∫–ª—é—á–µ–Ω–∏–µ ValueError
        lst = sorted([float(x) for x in lst])
    except ValueError:
        # –ï—Å–ª–∏ –∫–∞–∫–æ–π-–ª–∏–±–æ —ç–ª–µ–º–µ–Ω—Ç —Å–ø–∏—Å–∫–∞ –ø–æ–ª–Ω–æ—Å—Ç—å—é —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –±—É–∫–≤, –ª–∏–±–æ
        # –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø–æ–¥–∞–ª –Ω–∞ –≤—Ö–æ–¥ –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É, –æ—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏
        if any(filter(lambda x: x.isalpha(), lst)) or lst[0] == '':
            print('–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥')
            return None
    # –í—ã—á–∏—Å–ª—è–µ–º –º–µ–¥–∏–∞–Ω—É —Å–ø–∏—Å–∫–∞
    # –ï—Å–ª–∏ —Å–ø–∏—Å–æ–∫ –∏–º–µ–µ—Ç –Ω–µ—á—ë—Ç–Ω—É—é –¥–ª–∏–Ω—É, –Ω–∞—Ö–æ–¥–∏–º —Ü–µ–ª—É—é —á–∞—Å—Ç—å –æ—Ç –ø–æ–ª–æ–≤–∏–Ω—ã –¥–ª–∏–Ω—ã
    # –ó–Ω–∞—á–µ–Ω–∏–µ —Å –ø–æ–ª—É—á–µ–Ω–Ω—ã–º –∏–Ω–¥–µ–∫—Å–æ–º –∏ —è–≤–ª—è–µ—Ç—Å—è –º–µ–¥–∏–∞–Ω–æ–π
    if len(lst) % 2 == 1:
        print(f"Median: {lst[len(lst) // 2]}")
    # –ï—Å–ª–∏ —Å–ø–∏—Å–æ–∫ –∏–º–µ–µ—Ç —á—ë—Ç–Ω—É—é –¥–ª–∏–Ω—É, –Ω–∞—Ö–æ–¥–∏–º —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –º–µ–∂–¥—É –¥–≤—É–º—è
    # —Å—Ä–µ–¥–∏–Ω–Ω—ã–º–∏ —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏
    else:
        print(f"Median: {(lst[len(lst) // 2 - 1] + lst[len(lst) // 2]) / 2}")

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é
find_median(input_string)
```

–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã (—á–µ—Ä–µ–∑ —Ñ—É–Ω–∫—Ü–∏—é input()) –≤–≤–æ–¥–∏—Ç —Å—Ç—Ä–æ–∫—É ‚Äî –æ–ø–∏—Å–∞–Ω–∏–µ —á–∏—Å–ª–∞ —Å–ª–æ–≤–∞–º–∏ –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏—è:
–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî –Ω–æ–ª—å (0) –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ. –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ ‚Äî –æ–¥–∏–Ω –º–∏–ª–ª–∏–æ–Ω (1000000) –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.

–ü—Ä–∏ —Ä–µ—à–µ–Ω–∏–∏ –∑–∞–¥–∞—á–∏ –≤–∞–º –º–æ–∂–µ—Ç –ø—Ä–∏–≥–æ–¥–∏—Ç—å—Å—è –º–µ—Ç–æ–¥ —Å—Ç—Ä–æ–∫ s.startswith(word), –∫–æ—Ç–æ—Ä—ã–π –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è –ª–∏ —Å—Ç—Ä–æ–∫–∞ s —Å–æ —Å—Ç—Ä–æ–∫–∏ word. –ü–æ–¥—Ä–æ–±–Ω–µ–µ –æ –Ω—ë–º –≤—ã –º–æ–∂–µ—Ç–µ —É–∑–Ω–∞—Ç—å –≤ –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏.

```python
input_string = input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ —Å–ª–æ–≤–∞–º–∏: ')
# –≤–∞—à –∫–æ–¥ –∑–¥–µ—Å—å
# –ò—Å–ø–æ–ª—å–∑—É–µ–º –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–π –≤ —É—Å–ª–æ–≤–∏–∏ –∑–∞–¥–∞–Ω–∏—è —Å–ª–æ–≤–∞—Ä—å —Å —á–∏—Å–ª–∞–º–∏ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º–∏ –∏–º —Ñ—Ä–∞–≥–º–µ–Ω—Ç–∞–º–∏ –Ω–∞–∑–≤–∞–Ω–∏—è
number_word_dict = {
    "—Ç—ã": 1000, "–º–∏–ª–ª–∏–æ–Ω": 1000000,
    "—Å—Ç–æ": 100, "–¥–≤–µ—Å": 200, "—Ç—Ä–∏—Å": 300, "—á–µ—Ç—ã—Ä–µ—Å": 400, "–ø—è—Ç—å—Å": 500, "—à–µ—Å—Ç—å—Å": 600, "—Å–µ–º—å—Å": 700, "–≤–æ—Å–µ–º—å—Å": 800, "–¥–µ–≤—è—Ç—å—Å": 900,
    "–æ–¥–∏–Ω–Ω": 11, "–¥–≤–µ–Ω": 12, "—Ç—Ä–∏–Ω": 13, "—á–µ—Ç—ã—Ä–Ω": 14, "–ø—è—Ç–Ω": 15, "—à–µ—Å—Ç–Ω": 16, "—Å–µ–º–Ω": 17, "–≤–æ—Å–µ–º–Ω": 18, "–¥–µ–≤—è—Ç–Ω": 19,
    "–¥–≤–∞–¥": 20, "—Ç—Ä–∏–¥": 30, "—Å–æ—Ä": 40, "–ø—è—Ç—å–¥": 50, "—à–µ—Å—Ç—å–¥": 60, "—Å–µ–º—å–¥": 70, "–≤–æ—Å–µ–º—å–¥": 80, "–¥–µ–≤—è–Ω–æ": 90,
    "–¥–µ—Å": 10, "–Ω": 0, "–æ": 1, "–¥–≤": 2, "—Ç": 3, "—á": 4, "–ø": 5, "—à": 6, "—Å": 7, "–≤": 8, "–¥": 9, }

def transform_string_to_integer(s, num_dict):
    # –ï—Å–ª–∏ –≤–≤–µ–¥—ë–Ω–Ω–∞—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º —Å—Ç—Ä–æ–∫–∞ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —Å–ª–æ–≤,
    # —Ä–∞–∑–æ–±—å—ë–º –µ–µ –Ω–∞ —Å—Ç—Ä–æ–∫–∏ –∏ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫
    lst = [x for x in s.split(' ')] if ' ' in s else [s]
    # –°–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–æ–∫ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö —á–∏—Å–µ–ª
    num_list = []
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é, –≤ –∫–æ—Ç–æ—Ä—É—é –±—É–¥–µ—Ç –∑–∞–ø–∏—Å–∞–Ω –∫–æ–Ω–µ—á–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    result = 0
    # –°–æ–∑–¥–∞–¥–∏–º —Ü–∏–∫–ª –ø–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º lst
    # –§—É–Ω–∫—Ü–∏—è find_word_in_dict —Å–æ–æ—Ç–Ω–æ—Å–∏—Ç —Å–ª–æ–≤–æ –∏–∑ —Å–ø–∏—Å–∫–∞ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ –µ–º—É —á–∏—Å–ª–æ
    # –ó–∞–ø–æ–ª–Ω—è–µ–º —Å–ø–∏—Å–æ–∫ num_list –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω—ã–º–∏ —á–∏—Å–ª–∞–º–∏
    for i in lst:
        num_list.append(find_word_in_dict(i, num_dict))

    # –ï—Å–ª–∏ –≤ —Å–ø–∏—Å–∫–µ num_list —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —ç–ª–µ–º–µ–Ω—Ç, –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –µ–≥–æ –∫–∞–∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç,
    if len(num_list) == 1:
        result = num_list[0]
    else:
        # ...–≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ —Å–æ–∑–¥–∞—ë–º —Ü–∏–∫–ª –ø–æ –¥–ª–∏–Ω–µ —Å–ø–∏—Å–∫–∞ num_list - 1
        # –ï—Å–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤—ë–ª —á–∏—Å–ª–æ, —Å–æ—Å—Ç–æ—è—â–µ–µ –∏–∑ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —Å–ª–æ–≤, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã—á–∏—Å–ª–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç
        # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ current_elem –∏ next_elem –≤ —Ç–µ–ª–µ —Ü–∏–∫–ª–∞, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ
        # —Ç–µ–∫—É—â–µ–º—É –∏ –ø–æ—Å–ª–µ–¥—É—é—â–µ–º—É —ç–ª–µ–º–µ–Ω—Ç—É —Å–ø–∏—Å–∫–∞ num_list
        for i in range(len(num_list) - 1):
            if i == 0:
                current_elem = num_list[i]
            next_elem = num_list[i + 1]
            # –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é calc_num_elements
            # –ï—Å–ª–∏ —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ–Ω—å—à–µ –ø–æ—Å–ª–µ–¥—É—é—â–µ–≥–æ - –ø–µ—Ä–µ–º–Ω–æ–∂–∞–µ–º –∏—Ö,
            # –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ - —Å–∫–ª–∞–¥—ã–≤–∞–µ–º
            result = calc_num_elements(current_elem, next_elem)
            # –°–∫–ª–∞–¥—ã–≤–∞–µ–º –∏ –ø–µ—Ä–µ–º–Ω–æ–∂–∞–µ–º —ç–ª–µ–º–µ–Ω—Ç—ã —Å–ø–∏—Å–∫–∞ —Å–ª–µ–≤–∞ –Ω–∞–ø—Ä–∞–≤–æ
            current_elem = result
    
    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º —á–∏—Å–ª–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –¥–æ–ø—É—Å—Ç–∏–º–æ–º—É –¥–∏–∞–ø–∞–∑–æ–Ω—É,
    # —É–∫–∞–∑–∞–Ω–Ω–æ–º—É –≤ —É—Å–ª–æ–≤–∏–∏ –∑–∞–¥–∞–Ω–∏—è
    if result in range(0, 1000000):
        return result
    else:
        return "–ß–∏—Å–ª–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤–Ω–µ –¥–æ–ø—É—Å—Ç–∏–º–æ–≥–æ –¥–∏–∞–ø–∞–∑–æ–Ω–∞"

def find_word_in_dict(s, num_dict):
    # –°–æ–∑–¥–∞–¥–∏–º —Ü–∏–∫–ª –ø–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º —Å–ª–æ–≤–∞—Ä—è, –∏—Å–ø–æ–ª—å–∑—É–µ–º–æ–≥–æ –∫–∞–∫ –≤—Ö–æ–¥–Ω–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä —Ñ—É–Ω–∫—Ü–∏–∏
    for k, v in num_dict.items():
        # –ï—Å–ª–∏ –∫–ª—é—á —Å–ª–æ–≤–∞—Ä—è —è–≤–ª—è–µ—Ç—Å—è –Ω–∞—á–∞–ª–æ–º –≤–≤–µ–¥—ë–Ω–Ω–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º —Å—Ç—Ä–æ–∫–∏,
        if s.startswith(k):
            # ...–≤–µ—Ä–Ω—ë–º –∑–Ω–∞—á–µ–Ω–∏–µ, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ –Ω–∞–π–¥–µ–Ω–Ω–æ–º—É –∫–ª—é—á—É
            return v

calc_num_elements = lambda x1, x2: x1 + x2 if x1 > x2 else x1 * x2

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é
print(transform_string_to_integer(input_string, number_word_dict))
```

```python
from math import factorial
n = input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ: ')
# –≤–∞—à –∫–æ–¥ –∑–¥–µ—Å—å
def decompose_factorial(n):
    # –°–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–æ–∫ —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏, –ø–æ–ª—É—á–µ–Ω–Ω—ã–º–∏ –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏—è –≤–≤–µ–¥—ë–Ω–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º
    # —á–∏—Å–ª–∞ –Ω–∞ –ø—Ä–æ—Å—Ç—ã–µ –º–Ω–æ–∂–∏—Ç–µ–ª–∏
    multipliers = decompose(factorial(int(n)))
    # –í —Ü–µ–ª—è—Ö –Ω–∞—Å—Ç–æ—è—â–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è –∏–∑–º–µ–Ω–∏–º —ç–ª–µ–º–µ–Ω—Ç—ã —Å–ø–∏—Å–∫–∞ –Ω–∞ —Å—Ç—Ä–æ–∫–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
    multipliers = [str(x) for x in multipliers]

    # –ß—Ç–æ–±—ã —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å —Å—Ç–µ–ø–µ–Ω—å (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–≤—Ç–æ—Ä–æ–≤ –ø—Ä–æ—Å—Ç–æ–≥–æ –º–Ω–æ–∂–∏—Ç–µ–ª—è –≤ —Å–ø–∏—Å–∫–µ), —Å–æ–∑–¥–∞–¥–∏–º —Å–ª–æ–≤–∞—Ä—å,
    # –∫–ª—é—á–∞–º–∏ –±—É–¥—É—Ç —è–≤–ª—è—Ç—å—Å—è –∑–Ω–∞—á–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞, –∞ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–≤—Ç–æ—Ä–æ–≤
    multipliers_with_pows = {}
    # –°–æ–∑–¥–∞–¥–∏–º —Ü–∏–∫–ª –ø–æ –≤—Å–µ–º –∑–Ω–∞—á–µ–Ω–∏—è–º —Å–ø–∏—Å–∫–∞
    for i in multipliers:
        # –ï—Å–ª–∏ –ø—Ä–æ—Å—Ç–æ–≥–æ –º–Ω–æ–∂–∏—Ç–µ–ª—è –µ—â–µ –Ω–µ—Ç –≤ —Å–ª–æ–≤–∞—Ä–µ, –¥–æ–±–∞–≤–∏–º –µ–≥–æ
        if i not in multipliers_with_pows:
            multipliers_with_pows[i] = 1
        # –í –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –∏–Ω–∫—Ä–µ–º–µ–Ω—Ç–∏—Ä—É–µ–º —á–∏—Å–ª–æ –ø–æ–≤—Ç–æ—Ä–æ–≤
        else:
            multipliers_with_pows[i] += 1
    
    # –î–ª—è —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è –∫–æ–Ω–µ—á–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞ –∏—Å–ø–æ–ª—å–∑—É–µ–º –∏–º–µ—é—â–∏–π—Å—è –≤ —Ñ—É–Ω–∫—Ü–∏–∏ —Å–ø–∏—Å–æ–∫ multipliers
    multipliers = []
    # –°–æ–∑–¥–∞–¥–∏–º —Ü–∏–∫–ª –ø–æ –≤—Å–µ–º –∑–Ω–∞—á–µ–Ω–∏—è–º —Å–ª–æ–≤–∞—Ä—è
    for k, v in multipliers_with_pows.items():
        # –ï—Å–ª–∏ –ø—Ä–æ—Å—Ç–æ–π –º–Ω–æ–∂–∏—Ç–µ–ª—å –≤—Å—Ç—Ä–µ—á–∞–ª—Å—è –ª–∏—à—å –æ–¥–Ω–∞–∂–¥—ã (—Å—Ç–µ–ø–µ–Ω—å —á–∏—Å–ª–∞ —Ä–∞–≤–Ω–∞ 1), –¥–æ–±–∞–≤–∏—Ç—å –∫–ª—é—á —Å–ª–æ–≤–∞—Ä—è
        if v == 1:
            multipliers.append(k)
        # –ò–Ω–∞—á–µ –∑–∞–ø–∏—Å–∞—Ç—å —á–∏—Å–ª–æ –≤ –≤–∏–¥–µ n^k
        else:
            multipliers.append(f"{k}^{v}")
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—É—é —Å—Ç—Ä–æ–∫—É
    return f"{n}! = {' * '.join(multipliers)}"
    
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞–∑–ª–æ–∂–µ–Ω–∏—è —á–∏—Å–ª–∞ –Ω–∞ –ø—Ä–æ—Å—Ç—ã–µ –º–Ω–æ–∂–∏—Ç–µ–ª–∏
def decompose(n):
    # –°–æ–∑–¥–∞–¥–∏–º —Å–ø–∏—Å–æ–∫ –¥–ª—è –∫–æ–Ω–µ—á–Ω–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏—è
    result = []
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–≤—ã–π –º–Ω–æ–∂–∏—Ç–µ–ª—å
    d = 2
    # –°–æ–∑–¥–∞–¥–∏–º —Ü–∏–∫–ª —Å –ø–µ—Ä–µ–±–æ—Ä–æ–º –¥–æ n ** (1/2), –≤ —Ç–µ–ª–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –±—É–¥–µ–º –¥–µ–ª–∏—Ç—å –∏—Å—Ö–æ–¥–Ω–æ–µ —á–∏—Å–ª–æ
    # –Ω–∞ –ø—Ä–æ—Å—Ç—ã–µ –º–Ω–æ–∂–∏—Ç–µ–ª–∏, –Ω–∞—á–∏–Ω–∞—è —Å 2
    while d * d <= n:
        # –ï—Å–ª–∏ —á–∏—Å–ª–æ –¥–µ–ª–∏—Ç—Å—è –±–µ–∑ –æ—Å—Ç–∞—Ç–∫–∞, –¥–æ–±–∞–≤–ª—è–µ–º –≤ —Å–ø–∏—Å–æ–∫
        if n % d == 0:
            result.append(d)
            # –ò–∑–º–µ–Ω—è–µ–º –∏—Å—Ö–æ–¥–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º —Ü–µ–ª—É—é —á–∞—Å—Ç—å –æ—Ç –¥–µ–ª–µ–Ω–∏—è –Ω–∞ —Ç–µ–∫—É—â–∏–π –¥–µ–ª–∏—Ç–µ–ª—å
            n //= d
        # –ï—Å–ª–∏ —á–∏—Å–ª–æ –Ω–µ –¥–µ–ª–∏—Ç—Å—è –±–µ–∑ –æ—Å—Ç–∞—Ç–∫–∞, –∏–Ω–∫—Ä–µ–º–µ–Ω—Ç–∏—Ä—É–µ–º –¥–µ–ª–∏—Ç–µ–ª—å
        else:
            d += 1
    # –ï—Å–ª–∏ —á–∏—Å–ª–æ –±–æ–ª—å—à–µ –Ω–µ –¥–µ–ª–∏—Ç—Å—è –Ω–∞ –ø—Ä–æ—Å—Ç—ã–µ –º–Ω–æ–∂–∏—Ç–µ–ª–∏ –∏ –±–æ–ª—å—à–µ –µ–¥–∏–Ω–∏—Ü—ã - –¥–æ–±–∞–≤–∏–º –≤ —Å–ø–∏—Å–æ–∫
    if n > 1:
        result.append(n)
    return result

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é
decompose_factorial(n)
```

–ó–∞–¥–∞–Ω–∏–µ 6

–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã (—á–µ—Ä–µ–∑ —Ñ—É–Ω–∫—Ü–∏—é input()) –≤–≤–æ–¥–∏—Ç —Å—Ç—Ä–æ–∫—É, —Å–æ—Å—Ç–æ—è—â—É—é –∏–∑ —Å–∏–º–≤–æ–ª–æ–≤ –ª–∞—Ç–∏–Ω—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞.
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –∑–∞–º–µ–Ω—è—Ç—å –¥–≤–µ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –±—É–∫–≤—ã –≤–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –Ω–∞ —Å–ª–µ–¥—É—é—â—É—é –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É –±—É–∫–≤—É (–≤–∞–∂–Ω–æ: –∑–∞–º–µ–Ω—è–µ–º –ø–æ 2 –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö –±—É–∫–≤—ã –∑–∞ –æ–¥–Ω—É –∏—Ç–µ—Ä–∞—Ü–∏—é, –Ω–µ –±–æ–ª–µ–µ). –û–ø–µ—Ä–∞—Ü–∏—è –∑–∞–º–µ–Ω—ã –¥–æ–ª–∂–Ω–∞ –ø–æ–≤—Ç–æ—Ä—è—Ç—å—Å—è –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ –≤ —Å—Ç—Ä–æ–∫–µ –Ω–µ –æ—Å—Ç–∞–Ω—É—Ç—Å—è —Ç–æ–ª—å–∫–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã.
–ù–µ—Å–∫–æ–ª—å–∫–æ –≤–∞–∂–Ω—ã—Ö –º–æ–º–µ–Ω—Ç–æ–≤:
–±—É–∫–≤—ã –Ω–µ –æ–±—è–∑–∞–Ω—ã –±—ã—Ç—å —Å–º–µ–∂–Ω—ã–º–∏ (—Å–º–µ–∂–Ω—ã–µ ‚Äî —Å—Ç–æ—è—â–∏–µ —Ä—è–¥–æ–º; –≤ —Å—Ç—Ä–æ–∫–µ 'abc' –±—É–∫–≤—ã a –∏ b, –∞ —Ç–∞–∫–∂–µ b –∏ c —è–≤–ª—è—é—Ç—Å—è —Å–º–µ–∂–Ω—ã–º–∏, –≤ —Ç–æ –≤—Ä–µ–º—è –∫–∞–∫ a –∏ c ‚Äî –Ω–µ—Ç);
zz –∑–∞–º–µ–Ω—è–µ–º –Ω–∞ a;
–≤–≤–æ–¥–∏–º–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –±—É–∫–≤ –ª–∞—Ç–∏–Ω—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞ –≤ –Ω–∏–∂–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ.
–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é replace_duplicates(), –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—É—é —Å—Ç—Ä–æ–∫—É s, –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä–æ–π –≤–≤–æ–¥–∏—Ç—Å—è —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—å –æ–ø–∏—Å–∞–Ω–Ω—É—é –æ–ø–µ—Ä–∞—Ü–∏—é –∏ –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å —Å—Ç—Ä–æ–∫—É —Å –æ—Å—Ç–∞–≤—à–∏–º–∏—Å—è –±—É–∫–≤–∞–º–∏ –≤ –ª—é–±–æ–º –ø–æ—Ä—è–¥–∫–µ.

```python
import string

s = input('–í–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫—É: ')
# –≤–∞—à –∫–æ–¥ –∑–¥–µ—Å—å
def replace_duplicates(s):
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å—Ç—Ä–æ–∫—É –≤ —Å–ø–∏—Å–æ–∫
    s = [x for x in s]
    # –ü–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —à–∞–≥–∞ –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∫–∏ —Å—Ç—Ä–æ–∫–∏
    i = 0
    # –í —Ç–µ–ª–µ –æ—Å–Ω–æ–≤–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Å–æ–∑–¥–∞–¥–∏–º —Ü–∏–∫–ª, –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å—Å—è –¥–æ —Ç–µ—Ö –ø–æ—Ä, 
    # –ø–æ–∫–∞ –≤ —Å—Ç—Ä–æ–∫–µ –Ω–µ –∑–∞–∫–æ–Ω—á–∞—Ç—Å—è –¥—É–±–ª–∏–∫–∞—Ç—ã –±—É–∫–≤
    while len(s) > len(set(s)):
        # –¢—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Å—Ç—Ä–æ–∫—É
        s = change_string(s)
        i += 1
        # –í —Ü–µ–ª—è—Ö —Ä–∞—Å—à–∏—Ä–µ–Ω–∏—è –∑–∞–¥–∞–Ω–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –∫–∞–∂–¥—ã–π —à–∞–≥ –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∫–∏ —Å—Ç—Ä–æ–∫–∏
        print(f"step {i}: {''.join(s)}")
    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –∏–∑–º–µ–Ω—ë–Ω–Ω—É—é —Å—Ç—Ä–æ–∫—É, –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—É—é –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é
    return f"output: {''.join(sorted(s))}"

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –µ–¥–∏–Ω–∏—á–Ω–æ–≥–æ –∏–∑–º–µ–Ω–µ–Ω–∏—è –≤–≤–µ–¥—ë–Ω–Ω–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º —Å—Ç—Ä–æ–∫–∏
def change_string(s):
    # –°–æ–∑–¥–∞–¥–∏–º —Ü–∏–∫–ª –ø–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º —Å–ø–∏—Å–∫–∞, –∫—Ä–æ–º–µ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ
    for i, current_letter in enumerate(s[:-1]):
        # –î–æ—á–µ—Ä–Ω–∏–π —Ü–∏–∫–ª –ø—Ä–æ—Ö–æ–¥–∏—Ç –ø–æ –≤—Å–µ–º —ç–ª–µ–º–µ–Ω—Ç–∞–º —Å–ø–∏—Å–∫–∞, –∫—Ä–æ–º–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø–µ—Ä–≤–æ–≥–æ
        # –° –∫–∞–∂–¥—ã–º –ø–æ—Å–ª–µ–¥—É—é—â–∏–º —à–∞–≥–æ–º —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –ø—Ä–æ—Ö–æ–¥–∏—Ç —Ü–∏–∫–ª
        for letter in s[(i + 1):]:
            # –ï—Å–ª–∏ —Ç–µ–∫—É—â–∞—è –±—É–∫–≤–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ü–µ–ª–µ–≤–æ–π, –∏–Ω–∏—Ü–∏–∏—Ä—É–µ–º –∑–∞–ø–∏—Å—å —Å–ª–µ–¥—É—é—â–µ–π –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É –±—É–∫–≤—ã
            # –≤ –∫–æ–Ω–µ—Ü —Å—Ç—Ä–æ–∫–∏ –∏ —É–¥–∞–ª–µ–Ω–∏—é –∏–∑ —Å–ø–∏—Å–∫–∞ –∫–∞–∫ —Ç–µ–∫—É—â–µ–π –±—É–∫–≤—ã, —Ç–∞–∫ –∏ —Ü–µ–ª–µ–≤–æ–π (aa = b)
            if current_letter == letter:
                # –î–ª—è –≤–æ–∑–≤—Ä–∞—Ç–∞ –Ω–æ–≤–æ–π –±—É–∫–≤—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é get_next_letter
                s += get_next_letter(current_letter)
                s.remove(s[i])
                s.remove(current_letter)
                return s


# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤–æ–∑–≤—Ä–∞—Ç–∞ —Å–ª–µ–¥—É—é—â–µ–π –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É –±—É–∫–≤—ã
def get_next_letter(letter):
    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –∞–ª—Ñ–∞–≤–∏—Ç
    alphabet = string.ascii_lowercase
    # –°–æ–∑–¥–∞–¥–∏–º —Å–ª–æ–≤–∞—Ä—å, –∫–ª—é—á–∞–º–∏ –∫–æ—Ç–æ—Ä–æ–≥–æ –±—É–¥—É—Ç —è–≤–ª—è—Ç—å—Å—è –ø–æ—Ä—è–¥–∫–æ–≤—ã–µ –Ω–æ–º–µ—Ä–∞ –±—É–∫–≤ –≤ –∞–ª—Ñ–∞–≤–∏—Ç–µ,
    # –∞ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ - —Å–∞–º–∏ –±—É–∫–≤—ã
    alphabet_dict = {}
    for index, l in enumerate(alphabet):
        alphabet_dict[index + 1] = l
    # –°–æ–∑–¥–∞–¥–∏–º —Ü–∏–∫–ª, —á–µ—Ä–µ–∑ –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ–º –≤–æ–∑–≤—Ä–∞—â–∞—Ç—å –ø–æ—Å–ª–µ–¥—É—é—â—É—é –±—É–∫–≤—É
    for k, v in alphabet_dict.items():
        if letter == v:
            return alphabet_dict[k + 1]
        # –ï—Å–ª–∏ –Ω–∞ –≤—Ö–æ–¥ –ø–æ–¥–∞–Ω–∞ –ø–æ—Å–ª–µ–¥–Ω—è—è –±—É–∫–≤–∞ –∞–ª—Ñ–∞–≤–∏—Ç–∞ - –ø–µ—Ä–µ–¥–∞—Ç—å –ø–µ—Ä–≤—É—é –±—É–∫–≤—É
        elif letter == 'z':
            return 'a'

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é
print(replace_duplicates(s))
```

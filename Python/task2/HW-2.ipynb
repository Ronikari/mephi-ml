Задание 1

Пользователь с клавиатуры (через функцию input()) вводит через запятую и пробел две последовательности чисел (некоторые идентификаторы товаров/услуг, просмотренных клиентом 'A' и клиентом 'B' на платформе «Авито»).
Необходимо конвертировать каждый ввод пользователя в отдельный список, состоящий из целых чисел, а также составить новый список, состоящий из элементов пересечения двух исходных. Дубликаты необходимо удалить, а результирующий список — вывести на экран.
Важно! Если пересечение пустое, то на экран должен выводиться пустой список.
Если ввод пользователя невозможно конвертировать в список из целых чисел, возникает ошибка. Необходимо отлавливать эту ошибку и выводить на экран фразу Некорректный ввод.
Для поиска пересечений в списках напишите функцию pure_intersection(), которая будет возвращать список с уникальными элементами пересечения.

```python
input_string1 = input('Введите 1-ую последовательность идентификаторов: ')
input_string2 = input('Введите 2-ую последовательность идентификаторов: ')
# ваш код здесь
def pure_intersection(s1, s2):
    # Попытаемся сформировать списки из целых чисел
    try:
        # Удалим из первой строки запятые, разделим строку и запишем в список
        list1 = list(s1.replace(',', '').split(' '))
        # Преобразуем строковые значения в целые числа
        list1 = [int(x) for x in list1]
        # Проведём аналогичные действия для второй строки
        list2 = list(s2.replace(',', '').split(' '))
        list2 = [int(x) for x in list2]
    except ValueError:
        # Исключение возникает, если пользователь ввёл не целые числа
        print('Некорректный ввод')
        return
    # Преобразуем списки в множества и найдём их пересечения
    # Преобразуем конечный результат обратно в список
    result = list(set(list1).intersection(set(list2)))
    print(result)

# Инициализируем функцию
pure_intersection(input_string1, input_string2)
```

Задание 2

Пользователь подаёт на вход программе действительные числа через пробел с клавиатуры (через функцию input()).
Он привык пользоваться Excel, поэтому отделяет дробную часть числа запятой, а не точкой. Кроме того, он может случайно ввести число «словами» (например, не "5", а «пять») — в этом случае слово надо игнорировать.
Ваша задача — вывести минимум и максимум последовательности. Гарантируется, что пользователем будет введено хотя бы одно число.
Для нахождения минимума и максимума напишите функцию find_min_max, которая будет возвращать два числа.

```python
input_string = input('Введите последовательность чисел: ')
# ваш код здесь
def find_min_max(s):
    # Разделим строку и сформируем список
    lst = list(s.split(' '))
    try:
        # Попытаемся преобразовать все элементы списка в действительные числа
        lst = [float(x) for x in lst]
    except ValueError:
        # Если пользователь ввёл числа с запятой в качестве разделителя,
        # заменим на точку и добавим в список
        # Если пользователь ввёл значение, состоящее только из букв - проигнорируем
        lst = [float(x.replace(',', '.')) for x in lst if not x.isalpha()]
    # Вычислим минимальный и максимальный элементы списка
    print(f"Minimum: {min(lst)}")
    print(f"Maximum: {max(lst)}")

# Инициализируем функцию
find_min_max(input_string)
```

Задание 3

Пользователь с клавиатуры (через функцию input()) вводит через запятую и пробел неупорядоченную последовательность чисел.
Необходимо конвертировать ввод пользователя в список действительных чисел и вывести на экран медиану этой числовой последовательности.
Для поиска медианы в списке напишите функцию find_median(), которая будет возвращать одно число — медианное значение. Результат работы функции нужно вывести на экран.
Если переданная в функцию последовательность окажется пустой (пользователь ничего не ввёл) или пользователь ввёл числа некорректно (ввод нельзя преобразовать в числа), на экран должна быть выведена фраза Некорректный ввод.

```python
input_string = input('Введите последовательность чисел: ')
# ваш код здесь
def find_median(s):
    # Разделим строку и создадим список
    lst = list(s.replace(',', '').split(' '))
    try:
        # Попытаемся преобразовать элементы списка в действительные числа
        # и сортируем список по возрастанию
        # В противном случае выбрасываем исключение ValueError
        lst = sorted([float(x) for x in lst])
    except ValueError:
        # Если какой-либо элемент списка полностью состоит из букв, либо
        # пользователь подал на вход пустую строку, останавливаем выполнение функции
        if any(filter(lambda x: x.isalpha(), lst)) or lst[0] == '':
            print('Некорректный ввод')
            return None
    # Вычисляем медиану списка
    # Если список имеет нечётную длину, находим целую часть от половины длины
    # Значение с полученным индексом и является медианой
    if len(lst) % 2 == 1:
        print(f"Median: {lst[len(lst) // 2]}")
    # Если список имеет чётную длину, находим среднее арифметическое между двумя
    # срединными элементами
    else:
        print(f"Median: {(lst[len(lst) // 2 - 1] + lst[len(lst) // 2]) / 2}")

# Инициализируем функцию
find_median(input_string)
```

Задание 4

Пользователь с клавиатуры (через функцию input()) вводит строку — описание числа словами на русском языке.
Напишите функцию transform_string_to_integer(), которая принимает на вход введённую пользователем строку, конвертирует слова в число и возвращает число, которое описано в строке s.
Примечания:
Минимальное число — ноль (0) включительно. Максимальное число — один миллион (1000000) включительно.
Гарантируется, что на вход подаются строки с описанием числа, проверять это не нужно.

При решении задачи вам может пригодиться метод строк s.startswith(word), который определяет, начинается ли строка s со строки word. Подробнее о нём вы можете узнать в документации.

```python
input_string = input('Введите число словами: ')
# ваш код здесь
# Используем представленный в условии задания словарь с числами и соответствующими им фрагментами названия
number_word_dict = {
    "ты": 1000, "миллион": 1000000,
    "сто": 100, "двес": 200, "трис": 300, "четырес": 400, "пятьс": 500, "шестьс": 600, "семьс": 700, "восемьс": 800, "девятьс": 900,
    "одинн": 11, "двен": 12, "трин": 13, "четырн": 14, "пятн": 15, "шестн": 16, "семн": 17, "восемн": 18, "девятн": 19,
    "двад": 20, "трид": 30, "сор": 40, "пятьд": 50, "шестьд": 60, "семьд": 70, "восемьд": 80, "девяно": 90,
    "дес": 10, "н": 0, "о": 1, "дв": 2, "т": 3, "ч": 4, "п": 5, "ш": 6, "с": 7, "в": 8, "д": 9, }

def transform_string_to_integer(s, num_dict):
    # Если введённая пользователем строка состоит из нескольких слов,
    # разобьём ее на строки и инициализируем список
    lst = [x for x in s.split(' ')] if ' ' in s else [s]
    # Создадим список для полученных чисел
    num_list = []
    # Инициализируем переменную, в которую будет записан конечный результат
    result = 0
    # Создадим цикл по всем элементам lst
    # Функция find_word_in_dict соотносит слово из списка и возвращает соответствующее ему число
    # Заполняем список num_list преобразованными числами
    for i in lst:
        num_list.append(find_word_in_dict(i, num_dict))

    # Если в списке num_list только один элемент, возвращаем его как результат,
    if len(num_list) == 1:
        result = num_list[0]
    else:
        # ...в противном случае создаём цикл по длине списка num_list - 1
        # Если пользователь ввёл число, состоящее из нескольких слов, необходимо вычислить результат
        # Инициализируем переменные current_elem и next_elem в теле цикла, соответствующие
        # текущему и последующему элементу списка num_list
        for i in range(len(num_list) - 1):
            if i == 0:
                current_elem = num_list[i]
            next_elem = num_list[i + 1]
            # Для вычисления используем функцию calc_num_elements
            # Если текущее значение меньше последующего - перемножаем их,
            # в противном случае - складываем
            result = calc_num_elements(current_elem, next_elem)
            # Складываем и перемножаем элементы списка слева направо
            current_elem = result
    
    # Проверяем, что введённое пользователем число соответствует допустимому диапазону,
    # указанному в условии задания
    if result in range(0, 1000000):
        return result
    else:
        return "Число находится вне допустимого диапазона"

def find_word_in_dict(s, num_dict):
    # Создадим цикл по всем элементам словаря, используемого как входной параметр функции
    for k, v in num_dict.items():
        # Если ключ словаря является началом введённой пользователем строки,
        if s.startswith(k):
            # ...вернём значение, соответствующее найденному ключу
            return v

calc_num_elements = lambda x1, x2: x1 + x2 if x1 > x2 else x1 * x2

# Инициализируем функцию
print(transform_string_to_integer(input_string, number_word_dict))
```

Задание 5

Пользователь с клавиатуры (через функцию input()) вводит некоторое целое число n.
Напишите функцию decompose_factorial(), которая будет принимать на вход введённое пользователем число n, а затем вычислять факториал этого числа и выполнять разложение факториала на простые множители.
Напомним, что факториал числа  𝑛  ( 𝑛! ) — это произведение всех натуральных чисел от  1  до  𝑛  включительно.
Например, факториал числа  3  будет равен:  3!=1∗2∗3=6 .
Простое число — число, которое делится (нацело) только на себя и на 1. Например,  13  — простое число, так как оно делится только на себя (на  13 ) и на  1 , а  14  не является простым числом, так как оно делится на  7  и на  2 .
Подробнее о простых числах можно почитать здесь.
Простые множители — это простые числа (не считая 1), на которые число делится без остатка.
Например, для числа  10  простыми множителями являются числа  5  и  2 . Для числа  28  множителями являются числа  7  и  4 , но простыми множителями будут числа  7  и  2 , так как  4  — это  2∗2 , а число  28  представляется в виде  28=7∗22 .
В результате своей работы функция должна возвращать строку вида (приведён пример для n=12):
12! = 2^10 * 3^5 * 5^2 * 7 * 11
Здесь число слева от ^ — множитель, число справа от ^ — его степень. При этом если множитель встречается один раз, знак степени не выводится. Вывод множителей должен производиться в порядке их возрастания.

```python
from math import factorial
n = input('Введите число: ')
# ваш код здесь
def decompose_factorial(n):
    # Создадим список со значениями, полученными в результате разложения введённого пользователем
    # числа на простые множители
    multipliers = decompose(factorial(int(n)))
    # В целях настоящего задания изменим элементы списка на строковые значения
    multipliers = [str(x) for x in multipliers]

    # Чтобы сформировать степень (количество повторов простого множителя в списке), создадим словарь,
    # ключами будут являться значения списка, а значениями - количество повторов
    multipliers_with_pows = {}
    # Создадим цикл по всем значениям списка
    for i in multipliers:
        # Если простого множителя еще нет в словаре, добавим его
        if i not in multipliers_with_pows:
            multipliers_with_pows[i] = 1
        # В противном случае инкрементируем число повторов
        else:
            multipliers_with_pows[i] += 1
    
    # Для формирования конечного списка используем имеющийся в функции список multipliers
    multipliers = []
    # Создадим цикл по всем значениям словаря
    for k, v in multipliers_with_pows.items():
        # Если простой множитель встречался лишь однажды (степень числа равна 1), добавить ключ словаря
        if v == 1:
            multipliers.append(k)
        # Иначе записать число в виде n^k
        else:
            multipliers.append(f"{k}^{v}")
    # Возвращаем сформированную строку
    return f"{n}! = {' * '.join(multipliers)}"
    
# Функция для разложения числа на простые множители
def decompose(n):
    # Создадим список для конечного результата разложения
    result = []
    # Инициализируем первый множитель
    d = 2
    # Создадим цикл с перебором до n ** (1/2), в теле которого будем делить исходное число
    # на простые множители, начиная с 2
    while d * d <= n:
        # Если число делится без остатка, добавляем в список
        if n % d == 0:
            result.append(d)
            # Изменяем исходное значение и возвращаем целую часть от деления на текущий делитель
            n //= d
        # Если число не делится без остатка, инкрементируем делитель
        else:
            d += 1
    # Если число больше не делится на простые множители и больше единицы - добавим в список
    if n > 1:
        result.append(n)
    return result

# Инициализируем функцию
decompose_factorial(n)
```

Пользователь с клавиатуры (через функцию input()) вводит строку, состоящую из символов латинского алфавита.
Необходимо последовательно заменять две одинаковые буквы во введённой строке на следующую по алфавиту букву (важно: заменяем по 2 одинаковых буквы за одну итерацию, не более). Операция замены должна повторяться до тех пор, пока в строке не останутся только уникальные символы.
Несколько важных моментов:
буквы не обязаны быть смежными (смежные — стоящие рядом; в строке 'abc' буквы a и b, а также b и c являются смежными, в то время как a и c — нет);
zz заменяем на a;
вводимая строка состоит из букв латинского алфавита в нижнем регистре.
Напишите функцию replace_duplicates(), которая принимает на вход произвольную строку s, значение которой вводится с клавиатуры. Функция должна последовательно производить описанную операцию и возвращать строку с оставшимися буквами в любом порядке.

```python
import string

s = input('Введите строку: ')
# ваш код здесь
def replace_duplicates(s):
    # Преобразуем строку в список
    s = [x for x in s]
    # Переменная для определения шага корректировки строки
    i = 0
    # В теле основной функции создадим цикл, который будет выполняться до тех пор, 
    # пока в строке не закончатся дубликаты букв
    while len(s) > len(set(s)):
        # Трансформируем строку
        s = change_string(s)
        i += 1
        # В целях расширения задания возвращаем каждый шаг корректировки строки
        print(f"step {i}: {''.join(s)}")
    # Возвращаем изменённую строку, отсортированную по возрастанию
    return f"output: {''.join(sorted(s))}"

# Функция для единичного изменения введённой пользователем строки
def change_string(s):
    # Создадим цикл по всем элементам списка, кроме последнего
    for i, current_letter in enumerate(s[:-1]):
        # Дочерний цикл проходит по всем элементам списка, кроме целевого первого
        # С каждым последующим шагом уменьшается размерность, по которой проходит цикл
        for letter in s[(i + 1):]:
            # Если текущая буква соответствует целевой, инициируем запись следующей по алфавиту буквы
            # в конец строки и удалению из списка как текущей буквы, так и целевой (aa = b)
            if current_letter == letter:
                # Для возврата новой буквы используем функцию get_next_letter
                s += get_next_letter(current_letter)
                s.remove(s[i])
                s.remove(current_letter)
                return s


# Функция для возврата следующей по алфавиту буквы
def get_next_letter(letter):
    # Инициализируем алфавит
    alphabet = string.ascii_lowercase
    # Создадим словарь, ключами которого будут являться порядковые номера букв в алфавите,
    # а значениями - сами буквы
    alphabet_dict = {}
    for index, l in enumerate(alphabet):
        alphabet_dict[index + 1] = l
    # Создадим цикл, через который будем возвращать последующую букву
    for k, v in alphabet_dict.items():
        if letter == v:
            return alphabet_dict[k + 1]
        # Если на вход подана последняя буква алфавита - передать первую букву
        elif letter == 'z':
            return 'a'

# Инициализируем функцию
print(replace_duplicates(s))
```
